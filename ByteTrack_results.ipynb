```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
logResultsByteTrack = pd.read_csv("logResultsByteTrack.csv")
trackResultsByteTrack = pd.read_csv("trackResultsByteTrack.csv")
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].plot(list(range(1,80+1)), logResultsByteTrack["total_loss"], color='blue', label="total loss")
ax[0].set_title("Total Loss")
ax[0].set_xlabel("Epochs")
ax[0].set_ylabel("Loss")
ax[0].legend()

ax[1].plot(list(range(1,80+1)), logResultsByteTrack["iou_loss"], color='green', label="iou loss")
ax[1].plot(list(range(1,80+1)), logResultsByteTrack["conf_loss"], color='orange', label="confidence loss")
ax[1].plot(list(range(1,80+1)), logResultsByteTrack["cls_loss"], color='purple', label="classification loss")
ax[1].set_title("Losses")
ax[1].set_xlabel("Epochs")
ax[1].set_ylabel("Loss")
ax[1].legend()

ax[2].plot(list(range(1,80+1)), logResultsByteTrack["l1_loss"], color='red', label="l1 loss")
ax[2].set_title("L1 Loss")
ax[2].set_xlabel("Epochs")
ax[2].set_ylabel("Loss")
ax[2].legend()

plt.suptitle("ByteTrack MOT17half Training Loss", fontsize=18, fontweight="bold")

plt.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].plot(trackResultsByteTrack["epoch"], trackResultsByteTrack["IDF1"], color='blue')
ax[0].set_title("IDF1")
ax[0].set_xlabel("Epochs")
ax[0].set_ylabel("IDF1")

ax[1].plot(trackResultsByteTrack["epoch"], trackResultsByteTrack["Ids"], color='red')
ax[1].set_title("IDs")
ax[1].set_xlabel("Epochs")
ax[1].set_ylabel("IDs")

ax[2].plot(trackResultsByteTrack["epoch"], trackResultsByteTrack["MOTA"], color='green')
ax[2].set_title("MOTA")
ax[2].set_xlabel("Epochs")
ax[2].set_ylabel("MOTA")

plt.suptitle("ByteTrack MOT17 Validation Results", fontsize=18, fontweight="bold")

plt.tight_layout()

plt.show()
```
